```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Snapdrop_Aug 1, 2022 16_56_16.csv")
print(df.columns)
df.head()
```

```python
# Get numerical columns 
num_cols = [i for idx,i in enumerate(df.columns) if str(df.dtypes[idx]) in ['float64','int64']]
len(num_cols)
```

```python
# Plot stuff
fig = plt.figure(figsize=(10,8))

for i in range(16):
  plt.subplot(4,4, i+1)
  plt.xlabel(i)
  sns.distplot(df[num_cols[i]])
plt.tight_layout()
plt.show()
```
